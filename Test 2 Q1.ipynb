```python

def dydx(x, y):
    return (-2*x*y)

def rungeKutta(x0, y0, x, h):
    n = int((x - x0)/h)
    y=y0
    for i in range(1,n+1):
        k1 = h * dydx(x0, y)
        k2 = h * dydx(x0 + 0.5 * h, y + 0.5 * k1)
        k3 = h * dydx(x0 + 0.5 * h, y + 0.5 * k2)
        k4 = h * dydx(x0 + h, y + k3)
        y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4)
        
    x0=x0+h
    
    return y

x0 = 0
y = 1
x = 2
h = 0.2
print('The value of y at x is:', rungeKutta(x0, y, x, h))   

```

```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime

def get_weather_data(api_url):
    try:
        response = requests.get(api_url)
        response.raise_for_status()  # Raise an HTTPError for bad responses
        data = response.json()
        return data
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

def plot_temperature_trend(latitude, longitude, elevation, forecast_days, times, temperatures):
    plt.plot(times, temperatures, label='Temperature Trend', marker='o')
    plt.xlabel('Time')
    plt.ylabel('Temperature (°C)')
    plt.title(f'Temperature Trend for {forecast_days} Days\nLat: {latitude}, Lon: {longitude}, Elevation: {elevation}m')
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

def main():
    api_url = "https://api.open-meteo.com/v1/forecast?latitude=2.5&longitude=112.5&hourly=temperature_2m&timezone=Asia%2FSingapore&forecast_days="
    
    while True:
        try:
            forecast_days = int(input("Enter the number of forecast days (1-16): "))
            if 1 <= forecast_days <= 16:
                break
            else:
                print("Invalid input. Please enter a number between 1 and 16.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    weather_data = get_weather_data(api_url + str(forecast_days))

    if weather_data:
        latitude = weather_data['latitude']
        longitude = weather_data['longitude']
        elevation = weather_data['elevation']
        times = [datetime.strptime(timestamp, "%Y-%m-%dT%H:%M") for timestamp in weather_data['hourly']['time']]
        temperatures = weather_data['hourly']['temperature_2m']

        plot_temperature_trend(latitude, longitude, elevation, forecast_days, times, temperatures)

if __name__ == "__main__":
    main()
```
